# Meeting Minutes Voice Agent

Transcribes a meeting recording with a Hugging Face Whisper model, then uses OpenAI to turn the transcript into structured meeting minutes.

```python
import logging

from pipeline import (
    DEFAULT_AUDIO_MODEL,
    DEFAULT_TEXT_MODEL,
    load_credentials,
    transcribe_audio,
    generate_minutes,
)
from IPython.display import Markdown, display

logging.basicConfig(level=logging.INFO, format="%(asctime)s %(levelname)s %(message)s")

hf_token, openai_key = load_credentials()

AUDIO_FILE = "denver_extract.mp3"
```

## Step 1: Transcribe the audio with Hugging Face Whisper

```python
transcription = transcribe_audio(AUDIO_FILE, hf_token, model=DEFAULT_AUDIO_MODEL)

print(transcription)
```

## Step 2: Generate meeting minutes with OpenAI

```python
minutes = generate_minutes(transcription, openai_key, model=DEFAULT_TEXT_MODEL)
display(Markdown(minutes))
```
