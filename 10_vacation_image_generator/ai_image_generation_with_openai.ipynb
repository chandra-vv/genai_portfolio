# AI Image Generation with OpenAI

This notebook demonstrates how to use OpenAI's `gpt-image-1` model to generate images from text prompts. We explore different prompting strategies — from dynamic city-based prompts to fully custom scene descriptions — and display the results inline using PIL.

## 1. Setup & Imports

Import all required libraries: OpenAI SDK for image generation, Gradio for building a UI, PIL and `base64` for decoding and displaying the returned images.

```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
import base64
from io import BytesIO
from PIL import Image
```

## 2. Load API Key

Load the `OPENAI_API_KEY` from the `.env` file using `python-dotenv` and confirm it is available before making any API calls.

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
if openai_api_key:
    print(f"OpenAI API Key exists ")
else:
    print(f"OpenAI key not found")
```

## 3. Image Generator — Pop-Art Style

Defines the `artist(city)` function that calls `gpt-image-1` with a dynamic prompt requesting a **vibrant pop-art style** vacation scene for the given city. The API returns a base64-encoded image which is decoded into a PIL `Image` object.

```python
openai = OpenAI()
def artist(city):
    image_response = openai.images.generate(
            model="gpt-image-1",
            prompt=f"An image representing a vacation in {city}, showing tourist spots and everything unique about {city}, in a vibrant pop-art style",
            size="1024x1024",
            n=1,
        )
    image_base64 = image_response.data[0].b64_json
    image_data = base64.b64decode(image_base64)
    return Image.open(BytesIO(image_data))
```

### Generate Image — San Francisco (Pop-Art)

```python
image = artist("San Francisco City")
display(image)
```

## 4. Refined Prompt — No Style Constraint

Updated version of `artist(city)` with the pop-art style removed. The prompt now asks for a general vacation scene, giving the model more creative freedom over the visual style.

```python
def artist(city):
    image_response = openai.images.generate(
            model="gpt-image-1",
            prompt=f"An image representing a vacation in {city}, showing tourist spots and everything unique about {city}",
            size="1024x1024",
            n=1,
        )
    image_base64 = image_response.data[0].b64_json
    image_data = base64.b64decode(image_base64)
    return Image.open(BytesIO(image_data))
```

### Generate Image — San Francisco (No Style Constraint)

```python
image = artist("San Francisco city")
display(image)
```

## 5. Hardcoded Scene — Beach

Instead of a dynamic city input, this version uses a fully hand-crafted prompt describing a specific scene: a cinematic, photorealistic beach setting. This demonstrates how detailed prompt engineering gives more precise control over the output.

```python
def artist(city):
    image_response = openai.images.generate(
        model="gpt-image-1",
        prompt=(
            "A person relaxing on a beautiful sandy beach, "
            "with clear turquoise ocean waves, golden sunlight, "
            "palm trees swaying in the breeze, and a bright blue sky. "
            "Cinematic, photorealistic, warm and serene atmosphere."
        ),
        size="1024x1024",
        n=1,
    )
    image_base64 = image_response.data[0].b64_json
    image_data = base64.b64decode(image_base64)
    return Image.open(BytesIO(image_data))

image = artist("beach")
display(image)
```
